```python
import os
import pandas as pd
import json
import folium
import branca.colormap as cm
import numpy as np
import matplotlib.pyplot as plt

```

## DATA HANDLING

```python
# Loading the european map

country_geo_path = r'./topojson/europe.topojson.json'
topo_json_data = json.load(open(country_geo_path))
```

```python
# Loading the unemployement data set and 
unemployement_data_file=r'Eurostat_Table_unemployement.xls'
unemployement_data=pd.read_excel(unemployement_data_file,skiprows=3)
#unemployement_data_2016=unemployement_data['geo\\time']+unemployement_data['2016']
unemployement_data_2016=pd.concat([unemployement_data['geo\\time'][0:38], unemployement_data['2016'][0:38]], axis=1)
unemployement_data_2016=unemployement_data_2016[unemployement_data_2016['2016'] != ':']
unemployement_data_2016=unemployement_data_2016.apply(pd.to_numeric, errors='ignore')
unemployement_data_2016=unemployement_data_2016[4:]
display(unemployement_data_2016.head())
```

## Looking at the data

```python
quantiles = np.percentile(unemployement_data_2016['2016'],[25, 50, 75, 100])
quantiles.tolist()

#sortedData = unemployement_data_2016.sort_values(by= ['2016'])
#a = unemployement_data_2016['2016'].sort_values()
#quantiles = unemployement_data_2016['2016'].quantile(q=[0, .25, .5, .75, 1.], interpolation='linear')
```

```python
sortedEl = unemployement_data_2016['2016'].sort_values()
barlist=plt.bar(np.arange(len(sortedEl)) , sortedEl.iloc[np.arange(len(sortedEl))])
color = ['#e0ecf4','#9ebcda','#2b8cbe','#8856a7']
#quant25 = np.where(unemployement_data_2016['2016'] <= quantiles[0.25])[0]
for q,j in enumerate (quantiles):
    
    for i in np.where(sortedEl <= quantiles[3-q])[0]:
        barlist[i].set_color(color[3-q])
plt.show()
```

## Functions to color countries method I

```python
colormap = cm.linear.YlGn.scale(
    unemployement_data_2016['2016'].min(),
    unemployement_data_2016['2016'].max())

unemployment_dict = unemployement_data_2016.set_index('geo\\time')['2016']

linear = cm.LinearColormap(['green','yellow','red'], vmin=3., vmax=10.)

final=linear.to_step(
               data=unemployement_data_2016['2016'],
               method='quantiles',
               quantiles=[0, 0.3,0.6, 0.9, 1],
               round_method='int')
```

```python
def mycolor(pays):
    if unemployment_dict.get(pays,0)==0:
        return '#D3D3D3'
    else:
        color=final(unemployment_dict[pays])
        return color
```

```python
eu_map= folium.Map([48,15], tiles='cartodbpositron', zoom_start=4)
folium.TopoJson(topo_json_data,'objects.europe',
    style_function=lambda feature: {
        'fillColor': mycolor(feature['properties']['NAME']),
        'color': 'black',
        'fillOpacity':0.7,
        'weight': 1,
        'dashArray': '9, 10'}).add_to(eu_map)
final.caption = 'A colormap caption'
eu_map.add_child(final)

eu_map
```

## Functions to color countries method II

```python
def my_color_function(feature):
    """Maps low values to green and hugh values to red."""
    #a = np.array(unemployement_data_2016['2016'] > 10.5)
    #for i,j in enumerate(unemployement_data_2016['2016']): 
        #if a[i] == True:
    if not (feature in unemployment_dict.keys()):
        return '#969696'
    if unemployment_dict[feature] <= 5:    
        return '#ffffcc'
    elif (unemployment_dict[feature] > 5 and unemployment_dict[feature] < 10 ):
        return '#fed98e'
    elif (unemployment_dict[feature] > 10 and unemployment_dict[feature] < 15 ):
        return '#fd8d3c'
    else:
        return '#bd0026'
```
